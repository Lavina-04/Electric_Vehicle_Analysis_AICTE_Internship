```python
df = pd.read_csv("/content/Electric_Vehicle_Population_Size_History_By_County_.csv")
df.head()
```

```python
# Convert 'date' column to datetime format and extract year
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['year'] = df['date'].dt.year

# Check first few rows to confirm
df[['date', 'year']].head()
```

```python
# Group by year and count number of EVs
yearly = df.groupby('year')['electric_vehicle_(ev)_total'].sum().reset_index()

# Show the yearly EV count
print("\nEV count by year:")
print(yearly)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.plot(yearly['year'], yearly['electric_vehicle_(ev)_total'], marker='o')
plt.title("Electric Vehicles Growth by Year")
plt.xlabel("Year")
plt.ylabel("Number of Electric Vehicles")
plt.grid(True)
plt.show()
```

```python
# Save the cleaned and analyzed data
df.to_csv("Cleaned_Electric_Vehicle_Data.csv", index=False)

# Save your graph too (if you made one)
plt.savefig("EV_Trend_By_Year.png")

print(" Files saved successfully! You can now download them from the Files section on the left.")
```
